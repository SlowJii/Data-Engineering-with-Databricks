```python
# SPARK SESSION with DataFrame
from pyspark.sql import SparkSession

spark_session = SparkSession.builder \
    .appName("Spark Session Created") \
        .master("local[*]") \
            .config("spark.executor.memory", "4g") \
                .config("spark.driver.memory", "4g") \
                    .getOrCreate()

display(spark_session.range(1,10))
```

```python
# SPARK CONTEXT with RDD 
from pyspark import SparkConf,SparkContext

conf = SparkConf().setAppName("Spark Context Created") \
    .setMaster("local[*]") \
        .set("spark.executor.memory", "4g") \
            .set("spark.driver.memory", "4g")
            
sc = SparkContext(conf = conf)
data_sample_1 = sc.parallelize([1,2,3,4,5])
data_sample_2 = sc.parallelize([1,2,3,2,9,10])

result = data_sample_1.intersection(data_sample_2).collect()
```
