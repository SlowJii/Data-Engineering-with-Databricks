```python
%sql
/*
Để truy vấn file một cách chính xác nhất, syntax sẽ là 
SELECT * FROM file_format.`/path/to/file`
Phù hợp nhất cho hai dạng JSON và Parquet
*/

-- Extract dữ liệu dạng raw string
SELECT * FROM text.`path`;
-- Extract dữ liệu dạng binary, không cấu trúc
SELECT * FROM binaryFile.`path`;
SELECT * FROM csv.`path`;
SELECT * FROM json.`path`;
SELECT * FROM parquet.`path`;
```

```python
%sql
CREATE TABLE table_name AS SELECT * FROM file_format.`/path/to/file`
-- Câu lệnh CTAS tự động suy ra schema của file
```

```python
%sql
-- file_format không hỗ trợ tốt cho các dạng tệp csv,tsv, vậy có cách nào không
/*
Sử dụng USING

CREATE TABLE table_name
USING data_source
OPTIONS (key1 = val1, key2 = val2)
LOCATION `/path`

*/
-- Với cách tạo bảng trên, không thực sự tạo bảng vật lý mà tạo một External Table, trỏ đến nơi chứa dữ liệu chứ không thực sự chứa dữ liệu
-- Không phải bảng Delta
CREATE TABLE table_csv 
USING source.csv
OPTIONS (header='true', delimiter=',')
LOCATION 'path';


CREATE TABLE table_jdbc
USING jdbc
OPTIONS(url='jdbc:sqlite//anythings',
        dbtable='database.table',
        user='username',
        password='password'
);
```

```python
%sql
/*
External Table không phải là bảng Delta, vì vậy không thể đảm bảo hiệu suất một cách toàn diện và các tính năng ưu việt như quản lý thời gian, time traveling, __delta_log__ ..etc.. được

SOLUTION: Tạo một temp view trỏ thẳng đến file

CREATE TEMP VIEW temp_view_name(col1, col2)
USING data_source 
OPTIONS (key1 = val1, key2 = val2)
LOCATION `/path

CREATE TABLE table_name AS SELECT * FROM temp_view_name;
*/
-- Bằng cách trên, có thể dễ dàng tạo bảng Delta với dữ liệu bên ngoài
```
